```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn

from pylab import rcParams

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import scale
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2*np.random.randn(100,1)+3
rooms[0:10]
```

```python
price = 265 + 6*rooms + abs(np.random.randn(100,1))
price[0:10]
```

```python
plt.plot(rooms,price,'r^')

plt.xlabel('Number of Rooms, 2024 Avg')
plt.ylabel('2024 Avg Home Price,$1000s')

plt.show()
```

```python
x = rooms 
y = price

HotelLinReg = LinearRegression()
HotelLinReg.fit(x,y)
print(HotelLinReg.intercept_,HotelLinReg.coef_)

# Just Y=Mx+C
# M = coef_
# C = intercept_
```

```python
print(HotelLinReg.score(x,y))
```
