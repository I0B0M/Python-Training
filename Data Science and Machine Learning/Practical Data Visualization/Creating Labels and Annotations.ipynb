```python
%matplotlib  inline
rcParams['figure.figsize'] = 12,4
sns.set_style('darkgrid')
```

```python
x = range(1,10)
y = [1,2,3,4,0.5,4,3,2,1]

plt.bar(x,y)

plt.xlabel('x-axis label')
plt.ylabel('y-axis label')
```

```python
z = [1,2,3,4,0.5]

veh_type = ['bicycle','motorbike','car','van','bus']
pie_colors = ['gold','silver','red','powderblue','pink']

plt.pie(z,labels = veh_type,colors=pie_colors)
plt.show
```

```python
address = '/Users/ibm/Desktop/Python/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg', 'cyl','disp','hp', 'drat','wt', 'qsec','vs','am',	'gear','carb']

mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,1,1])

mpg.plot()

ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names,rotation=60, fontsize ='medium')
ax.set_title('Miles Per Gallon Of Cars')

ax.set_xlabel('Car Names')
ax.set_ylabel('MPG')
```
