```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas

from pandas import DataFrame
```

```python
data = {'names' : ['steve','john','richard','sarah','randy','micheal','julie'],
        'age' : [20,22,20,21,24,23,22],
        'gender': ['Male','Male','Male','Female','Male','Male','Female'],
        'rank' : [2,1,4,5,3,7,6]}

df = DataFrame(data)
df
```

```python
# Matplotlib Bar Chart

plt.bar(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')

plt.show()
```

```python
# Using Seaborn Barplot

plot = sns.barplot(data=df,x='names',y='age')
plot.set_title('Comparing Ages')
plt.show()
```

```python
# Matplotlib Line Chart

plt.plot(df['names'],df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.show
```

```python
# Seaborn Line Chart

plot = sns.lineplot(data=df,x='names',y='age')
plt.show
```

```python
# Matplotlib Pie Chart

plt.pie(df['age'],labels=df['names'])
plt.title('Comparing Ages')
plt.show
```

```python
# Seaborn Pie Chart

colours = sns.color_palette('pastel')[0:7]
plt.pie(df['age'],labels=df['names'],colors=colours)
plt.show
```
